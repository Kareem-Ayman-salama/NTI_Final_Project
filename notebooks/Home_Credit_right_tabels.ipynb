```python
import pandas as pd

# تحميل البيانات
credit_card_balance = pd.read_csv('credit_card_balance.csv')

credit_card_balance.head(10)
```

```python
credit_card_balance.shape
```

```python
credit_card_balance.info()
```

```python
credit_card_balance.isnull().sum()
```

```python
# أو حذف الصفوف التي تحتوي على قيم مفقودة
credit_card_balance.dropna(inplace=True)
```

```python
credit_card_balance.columns.tolist()
```

```python
credit_card_balance.shape
```

```python
credit_card_balance.describe()
```

```python
# تنفيذ التجميع
# تجميع البيانات بناءً على SK_ID_CURR (معرف العميل) وحساب المجموع، المتوسط، والعدد
aggregation = credit_card_balance.groupby('SK_ID_CURR').agg({
    'AMT_BALANCE': ['sum', 'mean'],           # حساب المجموع والمتوسط للرصيد
    'AMT_CREDIT_LIMIT_ACTUAL': ['sum', 'mean'],  # حساب المجموع والمتوسط للحد الائتماني الفعلي
    'AMT_DRAWINGS_ATM_CURRENT': ['sum', 'mean'], # حساب المجموع والمتوسط للمبالغ المسحوبة من أجهزة الصراف الآلي
    'AMT_DRAWINGS_CURRENT': ['sum', 'mean'],    # حساب المجموع والمتوسط للمبالغ المسحوبة
    'AMT_PAYMENT_CURRENT': ['sum', 'mean'],    # حساب المجموع والمتوسط للمدفوعات
    'AMT_RECEIVABLE_PRINCIPAL': ['sum', 'mean'], # حساب المجموع والمتوسط للمبالغ المستحقة الرئيسية
    'AMT_RECIVABLE': ['sum', 'mean'],         # حساب المجموع والمتوسط للمبالغ المستحقة
    'AMT_TOTAL_RECEIVABLE': ['sum', 'mean'],  # حساب المجموع والمتوسط للمبالغ المستحقة الكلية
    'MONTHS_BALANCE': 'count',                # حساب عدد الأشهر لكل عميل
}).reset_index()

# تعديل أسماء الأعمدة لتكون أكثر وضوحًا
aggregation.columns = ['_'.join(col).strip() for col in aggregation.columns.values]

# عرض النتيجة
print(aggregation.head())

# حفظ النتيجة في ملف CSV جديد
aggregation.to_csv('aggregated_credit_card_balance.csv', index=False)
```

```python
print(aggregation.isnull().sum())  # للتحقق من القيم المفقودة
```

```python
aggregation.dropna(inplace=True)  # حذف الصفوف التي تحتوي على قيم مفقودة
```

```python
import pandas as pd

# تحميل الملف
previous_app = pd.read_csv('previous_application.csv')
```

```python
print(previous_app.columns)
```

```python
previous_app_agg = previous_app.groupby('SK_ID_CURR').agg({
    'AMT_CREDIT': ['mean', 'sum', 'max'],
    'AMT_ANNUITY': ['mean', 'sum', 'max'],
    'AMT_APPLICATION': ['mean', 'max'],
    'AMT_GOODS_PRICE': ['mean', 'max'],
    'DAYS_DECISION': ['min', 'max', 'mean'],
    'CNT_PAYMENT': ['mean', 'sum'],
    'NAME_CONTRACT_STATUS': lambda x: x.mode()[0] if not x.mode().empty else None,  # الحالة الأكثر شيوعًا
    'NAME_CASH_LOAN_PURPOSE': lambda x: x.mode()[0] if not x.mode().empty else None  # الغرض الأكثر شيوعًا
}).reset_index()

# حفظ الجدول بعد التجميع في ملف CSV جديد
previous_app_agg.to_csv('previous_app_agg.csv', index=False)
print("تم حفظ البيانات بعد التجميع في ملف previous_app_agg.csv")
```

```python

import pandas as pd

# تحميل الملف
bureau_balance = pd.read_csv('bureau_balance.csv')
```

```python
bureau_balance.shape
```

```python
bureau_balance.info()
```

```python
bureau_balance.head()
```

```python
# إجراء عملية التجميع على البيانات بناءً على SK_ID_BUREAU
bureau_balance_agg = bureau_balance.groupby('SK_ID_BUREAU').agg({
    'MONTHS_BALANCE': ['min', 'max', 'mean'],  # إحصائيات عمود MONTHS_BALANCE
    'STATUS': 'count'                         # حساب عدد السجلات لكل SK_ID_BUREAU
}).reset_index()

# تسمية الأعمدة الجديدة لتوضيح الإحصائيات
bureau_balance_agg.columns = [
    'SK_ID_BUREAU',
    'MONTHS_BALANCE_MIN',
    'MONTHS_BALANCE_MAX',
    'MONTHS_BALANCE_MEAN',
    'STATUS_COUNT'
]

# عرض النتائج بعد التجميع
print(bureau_balance_agg.head())

# حفظ الجدول بعد التجميع في ملف CSV جديد
bureau_balance_agg.to_csv('bureau_balance_agg.csv', index=False)
print("تم حفظ البيانات بعد التجميع في ملف bureau_balance_agg.csv")
```

```python
import pandas as pd

# قراءة البيانات
bureau = pd.read_csv('bureau.csv')
print(bureau.head())
```

```python
# استخراج الأعمدة العددية فقط
numeric_cols = bureau.select_dtypes(include=['number']).columns
```

```python
# عملية التجميع
bureau_agg = bureau.groupby('SK_ID_CURR')[numeric_cols].agg(['mean', 'sum', 'count', 'std']).reset_index()

# تحسين أسماء الأعمدة
bureau_agg.columns = ['_'.join(col).strip('_') for col in bureau_agg.columns]
print(bureau_agg.head())

# حفظ النتائج في ملف CSV
bureau_agg.to_csv('bureau_agg.csv', index=False)
print("تم حفظ البيانات المجمعة في ملف 'bureau_agg.csv'")
```
